## skill 분석

```python
import pandas as pd

employ_final = pd.read_csv("employ_final.csv")
employ_final.head()
```

### 데이터 확인

```python
for i in employ_final["스킬"]:
    print(i)
```

### 쉼표, 공백으로 분류하기
- 쉼표로 분리된 string이 있고
- 공백으로 분리된 string이 있고
- 둘 다 함께 사용해서 분리된 string이 존재

```python
skill_type = []
for skill in employ_final["스킬"]:
    if type(skill) == float:
        continue

    if ',' in skill and ' ' in skill:
        for s in skill.split(','):
            
            skill_type.append(s)

    elif ',' in skill:
        for s in skill.split(','):
            skill_type.append(s)
            
    else:
        skill_type.append(skill)
        
skill_type
```

### /, \n 제거

```python
for idx, skill in enumerate(skill_type):
    skill_type[idx] = skill.strip()
    if '\n' in skill:
        for s in skill.split('\n'):
            skill_type.append(s)
            skill_type.pop(idx)
```

### 공백 제거

```python
for idx, skill in enumerate(skill_type):
    skill_type[idx] = skill.strip()
```

### 공백으로 단어 쪼개기

```python
for idx, skill in enumerate(skill_type):
    if ' ' in skill:
        for s in skill.split(' '):
            skill_type.append(s)
            skill_type.pop(idx)
```

### 분리된 기술 용어들이 얼마나 많이 등장했는지 파악

- 대략 191개정도
- 너무 많다

```python
skill_set= []

for skill in skill_type:
    if not skill in skill_set:
        skill_set.append(skill)

skill_set
```

```python
len(skill_set)
```

### skill 별로 몇 번 등장했는지 파악하기 위한 딕셔너리 선언

```python
skill_dic ={}
for i in skill_set:
    skill_dic[i] = 0
skill_dic
    
```

### 고유한 skill 들이 몇 번 등장했는지 파악
- 등장할 때 마다 해당 skill의 value 값을 1씩 증가시킴

```python

for skill in skill_type:
    skill_dic[skill] += 1
    
skill_dic
```

### 최종적으로 8번 이상 나타난 용어들만 시각화 하자

```python
for skill in skill_set:
    try:
        if skill_dic[skill] < 8:
            skill_dic.pop(skill)
    except:
        continue
skill_dic
        
```

```python
len(skill_dic)
```

### 시각화
- C++, C, Python, Linux, ROS등 단어들이 많이 등장했다

```python
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
x = []
y = []
for  i in skill_dic:
    x.append(i)
    y.append(skill_dic[i])
```

```python
skill_df = pd.DataFrame(x)
```

```python
skill_df["num"] = y
skill_df.sort_values("num", ascending=False)
```

### bar 그래프

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax = sns.barplot(x = 0, y="num", hue= 0, data=skill_df.sort_values("num", ascending=False))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height , height, ha = 'center', size = 15)
plt.title("가장 많이 등장한 기술 스택 단어 상위 15개")
plt.xlabel("기술 스택 단어")
plt.legend(loc = 'upper right')
plt.show();
```

### Pie 차트

```python
import matplotlib.pyplot as plt
import numpy as np
 
## 데이터 준비
labels = skill_df[0]## 라벨
frequency = skill_df["num"] ## 빈도
 
fig = plt.figure(figsize=(12,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 배경색을 하얀색으로 설정
ax = fig.add_subplot() ## 프레임 생성
 
pie = ax.pie(frequency, ## 파이차트 출력
       startangle=90, ## 시작점을 90도(degree)로 지정
       counterclock=False, ## 시계 방향으로 그린다.
       autopct=lambda p : '{:.2f}%'.format(p), ## 퍼센티지 출력
       wedgeprops=dict(width=0.5),  ## 중간의 반지름 0.5만큼 구멍을 뚫어준다.
       labels=skill_df[0] 
       )
plt.legend(pie[0],labels, loc= "upper right") ## 범례 표시
plt.show()
```
