## 01.기본 그래프 

```python
datas = [1, 2, 3, 4, 7, 4, 8, 9, 2, 3]
```

```python
# 가로 세로이고 단위는 inch
plt.figure(figsize=(3, 3))
plt.plot(datas)
plt.show()
```

```python
x_datas = np.arange(0, 200, 20)
x_datas
```

```python
y_datas = datas
```

```python
plt.figure(figsize=(5,5))
plt.plot(x_datas, y_datas)
plt.show()
```

```python
y_datas = np.random.randint(10, size=10)
```

```python
plt.figure(figsize=(5,5))
## plot은 선 그래프
# plt.plot(x_datas, y_datas)
## bar는 막대 그래프
# plt.bar(x_datas, y_datas)
## scatter는 점 그래프
plt.scatter(x_datas, y_datas)
## 제목과 x와 y축의 정보에 관해 설명
plt.title('점심은 뭐 목자')
plt.xlabel('구내식당?')
plt.ylabel('햄버거')
plt.show()
```

## 02. 스타일

```python
plt.figure(figsize=(3,3))
#r은 red ^ marker, -.은 linestyle
#디테일 할 필요는 없음
plt.plot(y_datas, 'r^-.')
plt.show()
```

```python
plt.figure(figsize=(3,3))
#c=color
#lw = line width
#ls = line style
#marker = 점을 표현하는 방법
#mec = marker의 컬러
#mew = marker의 width
#mfc = marker 내부의 color
plt.plot(y_datas, c='g', lw =5, ls=':', marker = '*', mec='r', mew = 1, mfc = 'y')
plt.show()
```

## 03.축 설정

```python
plt.figure(figsize=(5,2))
plt.plot(y_datas)
## x,y축 설정
plt.xlim(-2, 11)
plt.ylim(-2, 15)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(y_datas, 'r^-.')
plt.xticks(range(0,10))
plt.xticks(range(0, 11, 5))

plt.show()
```

## 04. 그래프 여러 개 그리기

```python
first = np.random.randint(10, size =5)
second = np.random.randint(10, size =5)
third = np.random.randint(10, size =5)
```

```python
plt.figure(figsize = (3,3))
## 같은 모양만 여러개 그릴 수 있음
## plot과 scater는 같이 그릴 수 없다
plt.scatter(first, second)
plt.plot(second)
plt.plot(third)
plt.show()
```

```python
datas = []

for _ in range(3):
    datas.append(np.random.randint(10, size = 10))
```

```python
datas
```

```python
datas = [np.random.randint(10, size = 10) for _ in range(3)]
datas
```

```python
plt.figure(figsize=(5,5))
plt.plot(datas[0], 'ro--',datas[1], 'b*--', datas[2], 'g^-')
# plt.plot(datas[1], 'b*--')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(datas[0], label = 'data1')
plt.plot(datas[1], label = 'data2')
plt.plot(datas[2], label = 'data3')

plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
# 10 개 길이
datas = [np.random.randint(10, size =10) for _ in range(10)]
```

```python
plt.figure(figsize=(5,5))
plt.plot(datas[0], label = 'data1')
plt.legend(loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize=(5,5))
# plt.plot(datas[0], label = 'data1')
for idx in range(10):
    plt.plot(datas[idx], label = f'datas{idx}')
plt.legend(loc = 'upper right')
plt.show()
```
