```python
from sklearn.linear_model import LinearRegression
```

- 인구수에 따라 필요한 cctv 갯수를 예측하기 위해 선형회귀를 함

```python
# 모델에게 줄 입력데이터와 출력데이터

x = data_result['인구수'].values.reshape(-1, 1)
y = data_result['소계'].values
```

```python
y
```

```python
# 회귀분석모델

model = LinearRegression()
```

```python
# 회귀분석모델 훈련
model.fit(x, y)
```

```python
data_result.describe()
```

```python
new_x = np.linspace(100000, 700000, 100)
predicted = model.predict(new_x.reshape(-1, 1))
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(data_result['인구수'], data_result['소계'])
plt.plot(new_x, predicted)
plt.grid()
plt.show()
```

```python
data_result['오차'] = data_result['소계'] - model.predict(x)
data_result.head(1)
```

```python
df_sort_f = data_result.sort_values(by='오차', ascending= False)
df_sort_t = data_result.sort_values(by='오차')
```

```python
from matplotlib.colors import ListedColormap
```

```python
# 구글에 color picker 검색하면 색상에 따른 16진수 코드를 얻을 수 있다
color_step = ['#e74c3c', '#16131f', '#cc107b', '#ebc107', '#07ebc8', '#07eb12']
```

```python
my_mcap = ListedColormap(color_step)
```

```python
plt.figure(figsize=(15,10))
## cmap은 내가 선택한 color로 원하는 데이터를 시각화를 할 수 있다
plt.scatter(data_result['인구수'], data_result['소계'], c=data_result['오차'], cmap = my_mcap)
plt.plot(new_x, predicted, 'r')

# cctv가 많은 곳 상위, 하위 5개 이름 표현
df_sort_f = data_result.sort_values(by='소계', ascending= False)
df_sort_t = data_result.sort_values(by='소계')
for idx in range(5):

    plt.text(df_sort_f['인구수'][idx]*1.02, df_sort_f['소계'][idx]*0.98, df_sort_f.index[idx], fontsize=15)
    plt.text(df_sort_t['인구수'][idx]*1.02, df_sort_t['소계'][idx]*0.98, df_sort_t.index[idx], fontsize = 15)

plt.title("인구 수 대비 CCTV 갯수", fontsize= 20)
plt.xlabel("인구 수")
plt.ylabel("CCTV 수")

plt.colorbar()
plt.grid()
plt.show()
```
